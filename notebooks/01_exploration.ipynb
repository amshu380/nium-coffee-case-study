# Coffee Data Exploration

Initial inspection of the three raw datasets for the Nium coffee case study.

```python
from pathlib import Path

import pandas as pd

project_root = Path.cwd().resolve().parent
raw_dir = project_root / 'data' / 'raw'

coffee_path = raw_dir / 'psd_coffee.csv'
population_path = raw_dir / 'world_population.csv'
country_codes_path = raw_dir / 'country_codes.csv'

print(f'Project root: {project_root}')
print(f'Raw data directory: {raw_dir}')
```

```python
coffee_df = pd.read_csv(coffee_path)

print('=== Coffee Data ===')
print('Shape:', coffee_df.shape)
print('\nColumns:')
print(coffee_df.columns.tolist())
print('\nHead(5):')
print(coffee_df.head(5))
print('\nDtypes:')
print(coffee_df.dtypes)
print('\nUnique Attribute_Description values:')
print(sorted(coffee_df['Attribute_Description'].dropna().unique().tolist()))
print('\nUnique Country_Name values:')
print(sorted(coffee_df['Country_Name'].dropna().unique().tolist()))
```

```python
population_df = pd.read_csv(population_path, skiprows=4)

print('=== Population Data ===')
print('Shape:', population_df.shape)
print('\nColumns:')
print(population_df.columns.tolist())
print('\nHead(5):')
print(population_df.head(5))
print('\nDtypes:')
print(population_df.dtypes)
print('\nUnique Country Name count:')
print(population_df['Country Name'].nunique(dropna=True))
year_columns = [col for col in population_df.columns if str(col).isdigit()]
print('\nLast 5 year columns:')
print(year_columns[-5:])
```

```python
country_codes_df = pd.read_csv(country_codes_path, sep=';')

print('=== Country Codes Data ===')
print('Shape:', country_codes_df.shape)
print('\nColumns:')
print(country_codes_df.columns.tolist())
print('\nHead(5):')
print(country_codes_df.head(5))
print('\nDtypes:')
print(country_codes_df.dtypes)
print('\nAll columns:')
print(country_codes_df.columns.tolist())

preferred_name_col = next((col for col in [
    'LABEL EN',
    'English short name lower case',
    'English short name',
    'CLDR display name',
    'official_name_en'
] if col in country_codes_df.columns), None)
iso2_col = next((col for col in [
    'ISO2 CODE',
    'Alpha-2 code',
    'ISO2',
    'iso2_code'
] if col in country_codes_df.columns), None)
iso3_col = next((col for col in [
    'ISO3 CODE',
    'Alpha-3 code',
    'ISO3',
    'iso3_code'
] if col in country_codes_df.columns), None)

sample_columns = [col for col in [preferred_name_col, iso2_col, iso3_col] if col is not None]
print('\nSample 10 rows with English name, ISO2, ISO3:')
print(country_codes_df[sample_columns].head(10))
```

```python
coffee_countries = sorted(coffee_df['Country_Name'].dropna().unique().tolist())
population_countries = sorted(population_df['Country Name'].dropna().unique().tolist())

country_entry_col = preferred_name_col or country_codes_df.columns[0]
country_entries = sorted(country_codes_df[country_entry_col].dropna().astype(str).unique().tolist())

suspicious_tokens = [
    'union', 'world', 'other', 'former', 'ussr', 'yugoslavia', 'eu', 'area',
    'total', 'states', 'islands', 'arab', 'africa', 'america', 'asia', 'europe'
]
suspect_countries = [
    name for name in coffee_countries
    if any(token in name.lower() for token in suspicious_tokens)
]

print('=== Summary ===')
print('Total unique country names in coffee data:', len(coffee_countries))
print('Total unique country names in population data:', len(population_countries))
print('Total unique country entries in country codes:', len(country_entries))
print('10 sample coffee country names that may not match standard ISO names:')
print(suspect_countries[:10])
```
